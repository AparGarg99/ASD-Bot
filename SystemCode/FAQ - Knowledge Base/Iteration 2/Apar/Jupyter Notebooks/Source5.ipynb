# Scrape QA pairs from website

```python
# Open webpage in a new window
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(link)
```

```python
# Parse text in webpage
source = driver.page_source
soup = bs4.BeautifulSoup(source, 'html.parser')
```

```python
x = soup.find('div',{'class':'article-rollup-container rollup-container'})
pages = []
for i in x.find_all('a'):
    print(i['href'])
    pages.append(i['href'])
```

```python
pages=pages[:-2]
```

```python
l1=[]
l2=[]
l3=[]
for url in pages:
    data = requests.get(url)
    soup2 = bs4.BeautifulSoup(data.text,'html.parser')
    
    ############################# TITLE OF PAGE ##########################
    title = soup2.find('h1',{'class':'page-title'}).text
    #print(title)
    #print('-'*50)
    l1.append(title)
    
    ############################# CONTENT ###############################
    t = soup2.text
    t = t.split('Page Content')[1]
    t = t.split('More information')[0]
    t = t.split('Additional')[0]
    t = t.strip()
    #print(t)
    #print('*'*100)
    l2.append(t)
    
    ############################# SUB HEADERS ###############################
    for i in soup2.find_all('h2'):
        k = i.text.strip()
        l3.append(k)  
```

```python
for k in l3:
    for j in range(len(l2)):
        if(k in l2[j]):
            l2[j] = l2[j].replace(k,'\n\n'+k)
```

```python
for i in range(len(l2)):
    print(i,l1[i])
    print('-'*50)
    print(l2[i])
    print('*'*100)
```

# Check which questions are similar

```python
stop_words = set(stopwords.words('english'))

def clean_text(sent):
    sent = sent.lower() # lowercase
    sent = re.sub(r'[^\w\s]', '', sent) # remove punctuations
    sent = re.sub('autism spectrum disorder','asd',sent) # Compress term
    sent = re.sub('autism','asd',sent) # Compress term
    sent = [w for w in sent.split() if not w.lower() in stop_words] # Remove stopwords
    sent = " ".join(sent)
    return sent
```

```python
def jaccard_similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return float(len(s1.intersection(s2)) / len(s1.union(s2)))
```

```python
def sim(li):
    for c1,i in enumerate(li):
        i2 = clean_text(i)
        list1 = i2.split()

        temp1 = 0.0
        temp2 = ''
        temp3 = ''
        temp4 = 0
        temp5 = 0

        for c2,j in enumerate(df['Question']):
            j2 = clean_text(j)
            list2 = j2.split()

            sim = jaccard_similarity(list1, list2)

            if(sim>temp1):
                temp1 = sim
                temp2 = j
                temp3 = i
                temp4 = c2
                temp5 = c1

        if(temp1>0):
            print('Website --> ',temp3,'(Index {})'.format(temp5))
            print('100 questions list --> ',temp2,'(Index {})'.format(temp4))
            print(temp1)
            print('*'*100)
```
